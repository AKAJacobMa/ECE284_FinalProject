```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
import json
import os
from sklearn.metrics import classification_report
```

```python
# svm training results

# read the JSON file
with open(os.path.join('training_records', 'svm_confusion_counts_train.json'), 'r') as file:
    data = json.load(file)

ypred = np.array(data['ypred'])
ytrue = np.array(data['ytrue'])


# Compute confusion matrix
cm = confusion_matrix(ytrue, ypred, labels=[1, 2, 3, 4])

# Define class names (can also be strings like ['A', 'B', 'C', 'D'])
class_names = ['Class 1', 'Class 2', 'Class 3', 'Class 4']

# Plot
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix for SVM Classifier on Training Data")
plt.tight_layout()
plt.show()
```

```python
# svm test results

# read the JSON file
with open(os.path.join('training_records', 'svm_confusion_counts_test.json'), 'r') as file:
    data = json.load(file)

ypred = np.array(data['ypred'])
ytrue = np.array(data['ytrue'])



# Compute confusion matrix
cm = confusion_matrix(ytrue, ypred, labels=[1, 2, 3, 4])

# Define class names (can also be strings like ['A', 'B', 'C', 'D'])
class_names = ['Class 1', 'Class 2', 'Class 3', 'Class 4']

# Plot
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix for SVM Classifier on Test Data")
plt.tight_layout()
plt.show()
```
