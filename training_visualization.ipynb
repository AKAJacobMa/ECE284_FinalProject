```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
import json
import os
from sklearn.metrics import classification_report
```

```python
# svm training results

# read the JSON file
with open(os.path.join('training_records', 'svm_confusion_counts_train.json'), 'r') as file:
    data = json.load(file)

ypred = np.array(data['ypred'])
ytrue = np.array(data['ytrue'])


# Compute confusion matrix
cm = confusion_matrix(ytrue, ypred, labels=[1, 2, 3, 4])

# Define class names (can also be strings like ['A', 'B', 'C', 'D'])
class_names = ['Baseline', 'Stress', 'Amusement', 'Meditation']

# Plot
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix for SVM Classifier on Training Data")
plt.tight_layout()
plt.show()
```

```python
def plot_classification_report_scores(ytrue, ypred, model_name, dataset_name):
    
    report = classification_report(ytrue, ypred, labels=[1, 2, 3, 4], output_dict=True)
    classes = ['Baseline', 'Stress', 'Amusement', 'Meditation']
    metrics = ['precision', 'recall', 'f1-score']

    
    scores = {metric: [report[str(i)][metric] for i in range(1, 5)] for metric in metrics}

    
    x = np.arange(len(classes))
    width = 0.25

    fig, ax = plt.subplots(figsize=(10, 6))
    rects1 = ax.bar(x - width, scores['precision'], width, label='Precision')
    rects2 = ax.bar(x, scores['recall'], width, label='Recall')
    rects3 = ax.bar(x + width, scores['f1-score'], width, label='F1-score')

    ax.set_ylabel('Score')
    ax.set_ylim([0, 1.05])
    ax.set_title(f'Per-Class Metrics for {model_name} on {dataset_name} Data')
    ax.set_xticks(x)
    ax.set_xticklabels(classes)
    ax.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()


# === SVM Train ===
with open(os.path.join('training_records', 'svm_confusion_counts_train.json'), 'r') as f:
    svm_train = json.load(f)
plot_classification_report_scores(svm_train['ytrue'], svm_train['ypred'], 'SVM', 'Training')

# === SVM Test ===
with open(os.path.join('training_records', 'svm_confusion_counts_test.json'), 'r') as f:
    svm_test = json.load(f)
plot_classification_report_scores(svm_test['ytrue'], svm_test['ypred'], 'SVM', 'Test')

# === RF Train ===
with open(os.path.join('training_records', 'rf_confusion_counts_train.json'), 'r') as f:
    rf_train = json.load(f)
plot_classification_report_scores(rf_train['ytrue'], rf_train['ypred'], 'Random Forest', 'Training')

# === RF Test ===
with open(os.path.join('training_records', 'rf_confusion_counts_test.json'), 'r') as f:
    rf_test = json.load(f)
plot_classification_report_scores(rf_test['ytrue'], rf_test['ypred'], 'Random Forest', 'Test')
```

```python
def plot_recall_radar(ytrue_pred_dicts, model_labels, title):
    labels = ['Baseline', 'Stress', 'Amusement', 'Meditation']
    num_labels = len(labels)

    angles = np.linspace(0, 2 * np.pi, num_labels, endpoint=False).tolist()
    angles += angles[:1]  # close the loop

    fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

    for ytrue, ypred, label in ytrue_pred_dicts:
        report = classification_report(ytrue, ypred, labels=[1, 2, 3, 4], output_dict=True)
        recalls = [report[str(i)]['recall'] for i in range(1, 5)]
        recalls += recalls[:1]  # close the loop
        ax.plot(angles, recalls, label=label)
        ax.fill(angles, recalls, alpha=0.1)

    ax.set_theta_offset(np.pi / 2)
    ax.set_theta_direction(-1)
    ax.set_thetagrids(np.degrees(angles[:-1]), labels)
    ax.set_ylim(0, 1)
    ax.set_title(title)
    ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
    ax.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
#  ytrue/ypred
with open(os.path.join('training_records', 'svm_confusion_counts_test.json')) as f:
    svm_test = json.load(f)
with open(os.path.join('training_records', 'rf_confusion_counts_test.json')) as f:
    rf_test = json.load(f)

# [ (ytrue, ypred, label), ... ]
comparison = [
    (svm_test['ytrue'], svm_test['ypred'], 'SVM (Test)'),
    (rf_test['ytrue'], rf_test['ypred'], 'Random Forest (Test)')
]

# 
plot_recall_radar(comparison, ['SVM', 'RF'], 'Recall Comparison per Class (Test Set)')
```

```python
from sklearn.metrics import accuracy_score

# 
with open("training_records/svm_confusion_counts_train.json") as f:
    svm_train = json.load(f)
with open("training_records/svm_confusion_counts_test.json") as f:
    svm_test = json.load(f)
with open("training_records/rf_confusion_counts_train.json") as f:
    rf_train = json.load(f)
with open("training_records/rf_confusion_counts_test.json") as f:
    rf_test = json.load(f)

# 
accs = {
    "SVM (Train)": accuracy_score(svm_train["ytrue"], svm_train["ypred"]),
    "SVM (Test)": accuracy_score(svm_test["ytrue"], svm_test["ypred"]),
    "RF (Train)": accuracy_score(rf_train["ytrue"], rf_train["ypred"]),
    "RF (Test)": accuracy_score(rf_test["ytrue"], rf_test["ypred"]),
}

# 
plt.figure(figsize=(8,5))
plt.bar(accs.keys(), accs.values(), color=["#88c", "#66b", "#6c6", "#4a4"])
plt.ylim([0, 1.05])
plt.ylabel("Accuracy")
plt.title("Model Accuracy Comparison (Train vs Test)")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
